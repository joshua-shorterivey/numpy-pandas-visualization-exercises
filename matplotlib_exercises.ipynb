## Use matplotlib to plot the following equation:

```python
import matplotlib.pyplot as plt
import math
```

#### You'll need to write the code that generates the x and y points.
#### Add an anotation for the point 0, 0, the origin.

```python
x = list(range(-10, 11))
```

```python
y = [(n ** 2  - n + 2) for n in x]
plt.plot(x,y)
plt.annotate('Origin', xy=(0, 0), xytext=(-1,10))
```

## Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

#### y = sqrt(x)

```python
x = list(range(1,100))
y = [math.sqrt(n) for n in x]
plt.plot(x,y)
```

#### y = x^3

```python
x = list(range(1, 11))
y = [n**3 for n in x]
plt.plot(x,y)
```

#### y = 2^x

```python
x = list(range(1, 11))
y = [2**n for n in x]
plt.plot(x, y)
```

#### y = 1 / ( x + 1 )

```python
x = list(range(1, 100))
y = [1 / (n+1) for n in x]
plt.plot(x,y)
```

## Combine the figures you created in the last step into one large figure with 4 subplots.

```python
x = list(range(1,100))
y = [math.sqrt(n) for n in x]
y2 = [n**3 for n in x]
y3 = [2**n for n in x]
y4 = [1 / (n+1) for n in x]

n_rows = 2
n_cols = 2
plt.figure(figsize=(14,6))
plt.subplot(n_rows, n_cols, 1)
plt.plot(x,y)
plt.title('Sqrt of x')

plt.subplot(n_rows, n_cols, 2)
plt.plot(x, y2)
plt.title('$x^3$')

plt.subplot(n_rows, n_cols, 3)
plt.plot(x, y3)
plt.title('2 ^ x')

plt.subplot(n_rows, n_cols, 4)
plt.plot(x, y4)
plt.title('1 / (x + 1)')
```

## Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
plt.figure(figsize=(14,6))

x = list(range(1,11))
y = [math.sqrt(n) for n in x]
plt.plot(x,y, label='$\sqrt{x}$')

x = list(range(1, 11))
y = [n**3 for n in x]
plt.plot(x,y, label=r'$x^3$')

x = list(range(1, 11))
y = [2**n for n in x]
plt.plot(x,y, label='$2^x$')

x = list(range(1, 11))
y = [1 / (n+1) for n in x]
plt.plot(x,y, label=r'$\frac{1}{x+1}$')

plt.legend(loc='upper left')
plt.title('Graph of Equations')
```
