### Curves to graph

#### y = 0n + 1 and label the curve "O(1)"

#### y = log(n) and label the curve "O(log n)"

#### y = n and label the curve "O(n)"

#### y = n âˆ— l o g ( n ) and label it "O(n log n)"

#### y = n^2 and label it "O(n^2)"

#### y = 2^n and label it "O(2^n)"

#### y = n! and label it "O(n!)"

#### y = n^n and label it "O(n^n)"

```python
n = list(range(1, 20))
y = [0*n + 1 for n in n]
y2 = [math.log(n) for n in n]
y3 = [n for n in n]
y4 = [n * math.log(n) for n in n]
y5 = [n ** 2 for n in n]
y6 = [2 ** n for n in n]
y7 = [math.factorial(n) for n in n]
y8 = [n ** n for n in n]

plt.figure(figsize=(20, 6))

plt.plot(n, y, label='$\mathcal{O}(1)$')
plt.plot(n, y3, label='$\mathcal{O}(\log n)$')
plt.plot(n, y3, label='$\mathcal{O}(n)$')
plt.plot(n, y4, label='$\mathcal{O}(n \log{} n)$')
plt.plot(n, y5, label='$\mathcal{O}(n^2)$')
plt.plot(n, y6, label='$\mathcal{O}(2^n)$')
plt.plot(n, y7, label='$\mathcal{O}(n!)$')
plt.plot(n, y8, label='$\mathcal{O}(n^n)$')

plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend()
plt.ylim(0,20)
```
